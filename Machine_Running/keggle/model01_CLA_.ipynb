```python
import pandas as pd
import numpy as np

df = pd.read_csv('./classification220926/train.csv')

x = df.iloc[:,:-1]
y = df.loc[:,['income']]
y
```

```python
x.drop(['no','fnlwgt','education','relationship','native-country','marital-status'],axis=1, inplace=True)
```

```python
df.loc[df['income'] == 1]
```

```python
from sklearn.preprocessing import LabelEncoder #레이블 인코딩

y1 = df.workclass
y3 = df.occupation
y4 = df.race
y5 = df.sex

le = LabelEncoder()

workclass = le.fit_transform(y1)
occupation = le.fit_transform(y3)
race = le.fit_transform(y4)
sex = le.fit_transform(y5)

x['workclass'] = workclass
x['occupation'] = occupation
x['race'] = race
x['sex'] = sex

x
```

```python
df.corr()
```

```python
df.value_counts('capital-gain')
```

```python
from sklearn.preprocessing import StandardScaler #분산이 1으로 스케일링
scaler = StandardScaler()
scaler1 = StandardScaler()

x.rename(columns= {'capital-gain':'capital_gain', 'capital-loss':'capital_loss'}, inplace=True)
scalex = x[['capital_gain']]
scalex1 = x[['capital_loss']]

capitalG = scaler.fit_transform(scalex)
capitalL = scaler1.fit_transform(scalex1)

x['capital_gain'] = capitalG
x['capital_loss'] = capitalL
```

```python
x
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split #트레인과 test 비율 나눔

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# 로지스틱 회귀 모델로 학습하기

```python
from sklearn.linear_model import LogisticRegression


lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)

print(lr.score(x_train, y_train))
print(lr.score(x_test, y_test))
```

## 의사결정트리로 학습

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42,criterion='gini')
dt.fit(x_train, y_train) #여기서 x_train은 이미 위에서 스케일링함

print(dt.score(x_train, y_train)) #훈련 데이터
print(dt.score(x_test, y_test)) #테스트 데이터
```

# y_pred 생성 후 비교

```python
from sklearn.metrics import accuracy_score

y_pred = dt.predict(x_test)

accuracy_score(y_test,y_pred)



```
