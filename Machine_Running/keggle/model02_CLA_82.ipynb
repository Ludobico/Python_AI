```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./classification220926/train.csv')
dftest = pd.read_csv('./classification220926/test.csv')

x = df.iloc[:,:-1]
x1_test = df.iloc[:, :-1]
y = df.loc[:,['income']]
y1_test = df.loc[:,['income']]
```

```python
x.drop(['no', 'education', 'native-country'], axis=1, inplace=True)
x1_test.drop(['no', 'education', 'fnlwgt', 'capital-gain', 'capital-loss', 'native-country'],
             axis=1, inplace=True)
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x['workclass'] = x['workclass'].astype('category')
x['marital-status'] = x['marital-status'].astype('category')
x['relationship'] = x['relationship'].astype('category')
x['race'] = x['race'].astype('category')
x['sex'] = x['sex'].astype('category')
x['workclass'] = x['workclass'].astype('category')
age = scaler.fit_transform(x[['age']])
x['age'] = age


x1_test['workclass'] = x1_test['workclass'].astype('category')
x1_test['marital-status'] = x1_test['marital-status'].astype('category')
x1_test['relationship'] = x1_test['relationship'].astype('category')
x1_test['race'] = x1_test['race'].astype('category')
x1_test['sex'] = x1_test['sex'].astype('category')
x1_test['workclass'] = x1_test['workclass'].astype('category')
```

```python
x = pd.get_dummies(x)
x1_test = pd.get_dummies(x1_test)
x
```

```python
from sklearn.model_selection import train_test_split #트레인과 test 비율 나눔
from sklearn.model_selection import cross_val_score  # k-fold cross validation

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression


lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)

print(lr.score(x_train, y_train))
print(lr.score(x_test, y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier


abc = AdaBoostClassifier(random_state=1,n_estimators = 2000,learning_rate = 1)
abc.fit(x_train, y_train) #여기서 x_train은 이미 위에서 스케일링함

print(abc.score(x_train, y_train)) #훈련 데이터
print(abc.score(x_test, y_test)) #테스트 데이터
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=1,learning_rate=0.1,n_estimators = 500)
gb.fit(x_train, y_train) #여기서 x_train은 이미 위에서 스케일링함

print(gb.score(x_train, y_train)) #훈련 데이터
print(gb.score(x_test, y_test)) #테스트 데이터
```

```python
from sklearn.ensemble import BaggingClassifier

bc = GradientBoostingClassifier(random_state=1,n_estimators=1000)
bc.fit(x_train, y_train) #여기서 x_train은 이미 위에서 스케일링함

print(bc.score(x_train, y_train)) #훈련 데이터
print(bc.score(x_test, y_test)) #테스트 데이터
```
