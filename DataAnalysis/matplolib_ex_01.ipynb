## 라인 플롯
* show()은 시각화 명령을 실제로 차트로 렌더링 하고 마우스 움직임 등의 이벤트를 기다리라는 지시이다.
* 가장 간단한 플롯으로 데이터가 시간,순서 등에 따라 어떻게 변화하는지 보여주기 위해 사용한다.
* 스타일 문자열은 color, marker, line style 의 순서로 지정한다. 만약 이 중에 일부가 생략되면 디폴트값이 적용된다.

```python
import matplotlib.pyplot as plt
```

```python
plt.title('line plot')
plt.plot([10,20,30,40],[1,4,9,16], 'm^-') #첫번째 x축 데이터, 두번째 y축 데이터 #세번째 스타일문자열 데이터가 하나면 y축데이터
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4.5,5,0.1) #배열 x생성, 범위:[-4.5,5],0.1씩 증가
y = 2*x**2 #수식을 이용해 배열 x에 대응하는 배열 y 생성
plt.plot(x,y,'mo-')
plt.show()
```

```python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2 #파란색
y2 = 5*x+30 #주황색
y3 = 4*x**2 + 10 #녹색
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2
y3 = np.tan(x) #걍만듬
y4 = np.cos(x) #걍만듬
y5 = np.sin(x) #걍만듬

plt.figure(1) #1번 그래프 창을 생성함
plt.plot(x,y1)

plt.figure(2) #2번 그래프 창을 생성함
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y2)

plt.figure(2)
plt.clf() #2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x,y1)

plt.figure(3) #걍만듬
plt.plot(y3) #걍만듬
plt.plot(y4) #걍만듬
plt.plot(y5) #걍만듬

plt.show()
```
