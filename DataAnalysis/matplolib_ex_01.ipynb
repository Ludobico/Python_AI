## 라인 플롯
* show()은 시각화 명령을 실제로 차트로 렌더링 하고 마우스 움직임 등의 이벤트를 기다리라는 지시이다.
* 가장 간단한 플롯으로 데이터가 시간,순서 등에 따라 어떻게 변화하는지 보여주기 위해 사용한다.
* 스타일 문자열은 color, marker, line style 의 순서로 지정한다. 만약 이 중에 일부가 생략되면 디폴트값이 적용된다.

````python
import matplotlib.pyplot as plt
````

````python
plt.title('line plot')
plt.plot([10,20,30,40],[1,4,9,16], 'm^-') #첫번째 x축 데이터, 두번째 y축 데이터 #세번째 스타일문자열 데이터가 하나면 y축데이터
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4.5,5,0.1) #배열 x생성, 범위:[-4.5,5],0.1씩 증가
y = 2*x**2 #수식을 이용해 배열 x에 대응하는 배열 y 생성
plt.plot(x,y,'mo-')
plt.show()
````

````python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2 #파란색
y2 = 5*x+30 #주황색
y3 = 4*x**2 + 10 #녹색
plt.plot(x, y1, label='y1 label')
plt.plot(x, y2, label='y2 label')
plt.plot(x, y3, label='y3 label')
plt.legend()
plt.show()
````

````python
x = np.arange(-5,5,0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2


plt.figure(1) #1번 그래프 창을 생성함
plt.plot(x,y1,label='y1 line')

plt.figure(2) #2번 그래프 창을 생성함
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y2,label='y2 line')
plt.legend()

plt.figure(2)
plt.clf() #2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x,y1)


plt.show()
````

## 스타일 문자열 약자 의미
* color c 선 색깔
* linewidth lw 선 굵기
* linestyle ls 선 스타일
* marker 마커 종류
* markersize ms 마커 크기
* markeredgecolor mec 마커 선 색깔
* markeredgewidth mew 마커 선 굵기
* markerfacecolor mfc 마커 내부 색깔

````python
# plt.rcParams['font.family'] = 'Malgun Gothic' #rcParams 전역적으로 지정
plt.title('라인 플롯(line plot)')
plt.plot([10,20,30,40],[1,4,9,16],c='k', lw=2, ls='--', marker='o', ms=15, mec='k', mew=5, mfc='r')
plt.xlim(0,50) #x축 최소,최댓값
plt.ylim(0,20) #y축 최소,최댓값
plt.show()
````

# 그래프 저장
* savefig(arg1, arg2)를 사용
* arg1: 경로와 함께 저장할 파일명
* arg2 : 해상도(dpi)를 나타내며 기본 값은 72이다.
* Ex) plt.savefig('C:\images\fig01.png', dpi = 100)
 * 사용 가능한 이미지 파일 확장자
* jpg, jpeg, png, svg, svgz, eps, pdf, pgf, ps, raw, rgba, tif, tiff 등


````python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # -기호 나오게 표시해줌

x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title('LaTeX, 문자열로 tick label 정의')
plt.plot(x,c)
plt.xticks([-np.pi, -np.pi/2,0,np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],['Low','Zero','High'])
plt.grid(True) #격자 표시, 기본값:False
plt.show()
````

````python
x1 = np.linspace(0.0, 5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)
y1 = np.cos(2* np.pi * x1) * np.exp(-x1)
y2 = np.cos(2* np.pi * x2)

plt.suptitle('main title')

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2, 'r.-')
plt.title('B tale of 2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() # plot간의 간격을 자동 조정
plt.show()
````

````python
fig, ax0 = plt.subplots() # 복수개의 subplot생성
ax1 = ax0.twinx() # twinx():x축을 공유하는 새로운 Axes객체 생성
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.set_xlabel("공유되는 x축")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
plt.show()
````

````python
#컨투어 플롯

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256

x = np.linspace(-3, 3, n) 
y = np.linspace(-3, 3, n)

XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=0.75, cmap='inferno') ##xx, yy는 입력값 zz는 출력값, alpha는 투명도
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
````

## np.mashgrid()
* meshgrid 명령은 사각형 영역을 구성하는 가로축의 점들과 세로축의 점을 나타내는 두 벡터를 인수로 받아서 이 사각형 영역을 이루는 조합을 출력한다. 결과는 그리드 포인트의 x값만을 표시하는 행렬과 y 값만을 표시하는 행렬 두 개로 분리하여 출력한다.

````python
import matplotlib.pyplot as plt
import numpy as np

irum = ['hong', 'kim', 'lee']
jumsu = [90, 70, 80]

n_data = len(irum) # 전체 데이터 수는 3
index = np.arange(n_data) # 배열 생성 (0, 1, 2)
print(index)

c=['gray', 'gray', 'gray']
plt.bar(index, jumsu, color = c, tick_label = irum)
plt.show()
````

````python
y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정

xlabel = ['국어', '영어', '수학']

plt.title("Bar Chart")
plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center',  label="점수")
plt.bar(x+0.4, y2, yerr=err2, width=0.3, color='blue', align='center',  label="반평균")

plt.xticks(x+0.4/2, xlabel)
plt.ylim(0, 100)

plt.xlabel("과목")
plt.ylabel("점수")

plt.legend()

plt.show()
````

````python
y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정

xlabel = ['국어', '영어', '수학']

plt.title("Barh Chart")

plt.barh(x, y1, height=0.3, xerr=err1, ecolor='yellow', color='red',  align='center', label="점수")
plt.barh(x+0.3, y2, height=0.3, xerr=err2, color='blue', align='center',  label="반평균")

plt.yticks(x+0.3/2, xlabel)

plt.xlim(0, 100)

plt.xlabel("점수")
plt.ylabel("과목")

plt.legend()

plt.show()
````

````python
import matplotlib.pyplot as plt

labels = ['개구리', '돼지', '개', '통나무']

sizes = [15.55, 30, 45, 9.45]

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

explode = (0, 0.1, 0, 0)

plt.title("Pie Chart")

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=90, counterclock=False)

plt.axis('equal') #원으로 만들어줌

plt.show()
````

## 이미지
* ```plt.imread()```
    * 로컬 이미지 주소를 문자열로 전달하면 이미지 파일을 읽어온다.
    * 이미지는 3차원 배열이라고 볼 수 있다.
    * imread()로 읽어온 이미지는 자동으로 numpy 객체로 변환된다.

* ```plt.imshow()```
    * numpy로 된 이미지를 시각화한다.

````python
img = plt.imread('cat.jpg')
print('{0}차원 이미지'.format(img.ndim))
print(img.shape) #(높이,넓이,색상값(채널)) #1은 흑백, 3은 컬러(rgb) 
plt.imshow(img)
plt.show()
````
