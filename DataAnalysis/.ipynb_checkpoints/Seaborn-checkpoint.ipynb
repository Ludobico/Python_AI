# seaborn
* Matplotlib을 기반으로 다양핚 색상 테마와 통계용 차트 등의 기능을 추가핚 시각화
패키지이다.
* 시각화 기능은 Matplotlib 패키지에 의졲하며 통계 기능은 Statsmodels 패키지에
의존한다.
* Seaborn을 임포트하면 색상 등을 Matplotlib에서 제공하는 기본 스타일이 아닌
Seaborn에서 지정한 기본 스타일로 바꾼다.
* Seaborn은 스타일 지정을 위핚 색상 팔렛트(color palette)라는 것을 지원핚다. 색상
팔렛트는 Matplotlib의 칼라맵(colormap)으로 사용할 수도 잇다.
* Ex1) current_palette = sns.color_palette() 
 sns.palplot(current_palette)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # -기호 나오게 표시해줌

# titanic = sns.load_dataset('titanic') #타이타닉호 데이터
# tips = sns.load_dataset('tips') #팁 데이터
# flights = sns.load_dataset('flights') #여객운송 데이터
```

```python
iris = sns.load_dataset('iris') #붓꽃 데이터(분류용)
iris
```

```python
iris.info()
```

```python
iris.value_counts('species')
```

```python
x = iris.petal_length.values
sns.rugplot(x) 
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot") 
plt.show()
```

```python
x = iris.petal_length.values
sns.kdeplot(x) 
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot") 
plt.show()
```

```python
x = iris.petal_length.values
sns.distplot(x, kde=True, rug=True) 
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot") 
plt.show()
```

## 다차원 데이터
* 데이터 변수가 여러개인 다차원 데이터는 데이터의 종류에 따라 다음과 같은 경우가
잇을 수 잇다.
    * 분석하고자 하는 데이터가 모두 실수 값인 경우
    * 분석하고자 하는 데이터가 모두 카테고리 값인 경우
    * 분석하고자 하는 데이터가 모두 실수 값과 카테고리 값이 섞여 잇는 경우
* 2차원 실수형 데이터
     * jointplot 사용
* 다차원 실수형 데이터
     *pairplot 사용
* 2차원 카테고리 데이터
     *heatmap 사용
* 2차원 복합 데이터
     *barplot, boxplot, pointplot, violinplot, stripplot, swarmplot 사용

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='scatter')
plt.rcParams["font.family"] = 'Malgun Gothic' # 한글처리
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02) 
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='kde', color='g')
plt.rcParams["font.family"] = 'Malgun Gothic' # 한글처리
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02) 
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='hex', color='g')
plt.rcParams["font.family"] = 'Malgun Gothic' # 한글처리
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02) 
plt.show()
```

```python
sns.pairplot(iris) 
plt.title("Iris Data의 Pair Plot") 
plt.show()
```

```python
iris.corr()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"]) 
plt.title("Iris Data의 Pair Plot, Hue로 꽃의 종을 시각화", x=-1.3, y = 4.3) 
plt.show()
```

```python
titanic = sns.load_dataset('titanic')

titanic_size = titanic.pivot_table(index="class", columns="sex",  aggfunc="size")
sns.heatmap(titanic_size, cmap=sns.light_palette("gray", as_cmap=True),  annot=True, fmt="d")#annot 데이터를 표시하겠다, fmt 데이터타입
plt.title("Heatmap")
plt.show()
```

```python
flights = sns.load_dataset('flights')

flights_passengers = flights.pivot("month", "year", "passengers") 
plt.title("연도, 월 별 승객수에 대한 Heatmap") 
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1) 
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.barplot(x="day", y="total_bill", data=tips) 
plt.title("요일 별, 전체 팁")
plt.show()
```

## boxplot 중요하다함
* 박스-휘스커 플롯(Box-Whisker Plot) 혹은 갂단히 박스 플롯이라 부르는 차트를
그려준다.
* 박스 플롯은 박스와 박스 바깥의 선(whisker)으로 이루어짂다.
* 박스는 실수 값 분포에서 1사분위수(Q1)와 3사분위수(Q3)를 뜻하고 이 3사분위수와
1사분수의 차이(Q3 - Q1)를 IQR(interquartile range)라고 핚다.
* 박스 내부의 가로선은 중앙값을 나타낸다. 박스 외부의 세로선은 1사분위 수보다 1.5 x 
IQR 맊큼 낮은 값과 3사분위 수보다 1.5 x IQR 맊큼 높은 값의 구갂을 기준으로 그
구갂의 내부에 잇는 가장 큰 데이터와 가장 작은 데이터를 잆는 선분이다. 
* 그 바깥의 점은 아웃라이어(outlier)라고 부르는데 일일히 점으로 표시한다.

```python
 sns.boxplot(x="day", y="total_bill", data=tips) 
 plt.title("요일 별 전체 팁의 Box Plot") 

 plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips) 
plt.title("요일 별 전체 팁의 Violin Plot") 
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True) 
plt.title("요일 별 전체 팁의 Strip Plot") 
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips) 
plt.title("요일 별 젂체 팁의 Swarm Plot") 
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips) 
plt.title("요일 별, 성별 전체 팁의 Histogram") 
plt.show()
```

```python
data = titanic[titanic.survived.notnull()] 
sns.catplot(x="age", y="sex", hue="survived", row="class", data=data, kind="violin", split=True, height=2, aspect=4) 
plt.title("각 클래스의 성별 생존자 수의 Catplot") 
plt.show()
```

```python
import numpy as np
plt.title("Boxplot과 Strip Plot로 표현한, 요일 별 팁")
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf) 
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4") 
plt.show()
```
